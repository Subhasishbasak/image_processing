```python
def normalize_patch(image, x0, y0, patch_width):
    """
    Find the patch_width x patch_width subimage centered at x0,y0.
    Compute the mean and standard deviation of intensity values.
    Normalize intensity values over the patch.
    """
    center_x = math.floor(x0)
    center_y = math.floor(y0)
    w = patch_width // 2
    subImg = image[center_x-w:center_x+w, center_y-w:center_y+w]
    mu = subImg.mean()
    sigma = subImg.std()
    norm_patch = (subImg-mu)/sigma
    return norm_patch
```

```python
def get_features(image, x, y, feature_width, scales=None):
    k = min(x.size, y.size)
    fv = []
    for i in range(k):
        feature = normalize_patch(image, x[i], y[i], feature_width)
        fv.append(feature)
    return fv
```
